```python
import numpy as np
import pandas as pd

# Diccionario: columna -> categoría general
mapa_categorias = {
    "P1121_¿En cuanto a los siguientes aspectos, ¿Cuáles son las tres principales razones por la que considera a TransMilenio inseguro o muy inseguro?  /1. Estaciones, portales, paraderos o buses muy llenos": "Congestión / Operación",
    
    "P1122_Presencia de habitantes de calle": "Desorden social",
    "P1123_Presencia de vendedores ambulantes, cantantes, personas pidiendo dinero, etc.": "Desorden social",
    "P1126_Colados": "Desorden social",
    "P1128_Protestas, bloqueos, vandalismo": "Desorden social",
    
    "P1124_Actividades ilegales alrededor de las estaciones": "Delincuencia directa",
    "P1127_Peleas, riñas o intolerancias al interior de la estación y los buses": "Delincuencia directa",
    "P11211_Hurtos (atraco, cosquilleo, raponazo)": "Delincuencia directa",
    "P11212_Acoso sexual, manoseos , piropos, miradas o actos obscenos": "Delincuencia directa",
    
    "P1125_Espacios abandonados o deteriorados cerca a las estaciones.": "Infraestructura deteriorada",
    "P11210_Estaciones deterioradas (dañadas, vandalizadas, con poca iluminación o sin puertas)": "Infraestructura deteriorada",
    
    "P1129_Falta de presencia policial o de personal de seguridad": "Falta de control institucional"
}
```

```python
def agrupar_respuestas(row):
    categorias = set()
    
    for col, categoria in mapa_categorias.items():
        if pd.notna(row[col]):  # si la persona marcó esa opción
            categorias.add(categoria)
    
    if len(categorias) == 0:
        return np.nan
    
    return ", ".join(categorias)


df_EPV["P1121_categoria_general"] = df_EPV.apply(agrupar_respuestas, axis=1)
```

```python
mapa_codigos = {
    "Delincuencia directa": 1,
    "Desorden social": 2,
    "Infraestructura deteriorada": 3,
    "Falta de control institucional": 4,
    "Congestión / Operación": 5
}

df_EPV["P1121_¿En cuanto a los siguientes aspectos, ¿Cuáles son las tres principales razones por la que considera a TransMilenio inseguro o muy inseguro?"] = df_EPV["P1121_categoria_general"].map(mapa_codigos)
```

```python
import pandas as pd

df = df_EPV.copy()

id_col = "ID"
peso = "FACTOR"

def convertir_a_columna_unica(df, col_inicio, col_fin):
    
    idx_inicio = df.columns.get_loc(col_inicio)
    idx_fin = df.columns.get_loc(col_fin)
    cols = df.columns[idx_inicio:idx_fin+1]
    
    # Extraer encabezado común (texto entre "_" y el número)
    primer_nombre = cols[0]
    encabezado = primer_nombre.split("_", 1)[1]
    encabezado = encabezado.rsplit("  ", 1)[0]  # quitar "1.Saludo amable"
    
    temp = df[[id_col, peso] + list(cols)]
    
    df_long = temp.melt(
        id_vars=[id_col, peso],
        value_vars=cols,
        var_name="col_original",
        value_name="valor"
    )
    
    # Solo donde marcaron (Si es tipo Sí/No)
    df_long = df_long[df_long["valor"] == "Sí"]
    
    # Extraer solo la opción (1.Saludo amable)
    df_long["opcion"] = df_long["col_original"].str.split(":").str[-1].str.strip()
    
    # Renombrar la columna con el encabezado real
    df_long = df_long[[id_col, "opcion", peso]]
    df_long = df_long.rename(columns={"opcion": encabezado})
    
    return df_long
```

```python
import pandas as pd
import numpy as np

df= df_EPV.copy()

# ==========================================================
# 1️⃣ DEFINIR LOS RANGOS
# ==========================================================

rangos = [
    ("P2042221_", "P2042224_"),
    ("P204111_", "P204118_"),
    ("P204118_", "P204119_"),
    ("P2081_", "P2087_"),
    ("P204121_", "P204124_"),
    ("P4501_", "P4506_"),
    ("P4511_", "P4517_"),
    ("P4521_", "P45211_"),
    ("P4071_", "P4073_"),
    ("P4121_", "P4126_"),
]

# ==========================================================
# 2️⃣ FUNCIÓN PARA OBTENER COLUMNAS POR RANGO
# ==========================================================

def obtener_columnas_por_rango(df, inicio, fin):
    cols = df.columns.tolist()
    idx_inicio = next(i for i, c in enumerate(cols) if c.startswith(inicio))
    idx_fin = next(i for i, c in enumerate(cols) if c.startswith(fin))
    return cols[idx_inicio:idx_fin+1]


# ==========================================================
# 3️⃣ OPCIÓN 1: UNA SOLA RESPUESTA (PRIMERA NO NULA)
# ==========================================================

df_opcion1 = df.copy()

for inicio, fin in rangos:
    cols_rango = obtener_columnas_por_rango(df, inicio, fin)
    
    nueva_col = cols_rango[0].split("_",1)[1]  # texto pregunta
    
    df_opcion1[nueva_col] = df_opcion1[cols_rango] \
        .apply(lambda row: next((col.split("_",1)[1] 
                                 for col, val in row.items() 
                                 if pd.notnull(val) and val not in ["No", 0]), 
                                np.nan), axis=1)
    
    df_opcion1.drop(columns=cols_rango, inplace=True)


# ==========================================================
# 4️⃣ OPCIÓN 2: CONCATENAR RESPUESTAS
# ==========================================================

df_opcion2 = df.copy()

for inicio, fin in rangos:
    cols_rango = obtener_columnas_por_rango(df, inicio, fin)
    
    nueva_col = cols_rango[0].split("_",1)[1]
    
    def concatenar(row):
        respuestas = []
        for col, val in row.items():
            if pd.notnull(val) and val not in ["No", 0]:
                respuestas.append(col.split("_",1)[1])
        if len(respuestas) == 0:
            return np.nan
        return ", ".join(respuestas)
    
    df_opcion2[nueva_col] = df_opcion2[cols_rango].apply(concatenar, axis=1)
    
    df_opcion2.drop(columns=cols_rango, inplace=True)


# ==========================================================
# 5️⃣ OPCIÓN 3: FORMATO LONG (REPETIR REGISTROS)
# ==========================================================

dfs_long = []

for inicio, fin in rangos:
    cols_rango = obtener_columnas_por_rango(df, inicio, fin)
    
    df_melt = df.melt(
        id_vars=[col for col in df.columns if col not in cols_rango],
        value_vars=cols_rango,
        var_name="Pregunta_Opcion",
        value_name="Valor"
    )
    
    df_melt = df_melt[
        df_melt["Valor"].notna() & 
        (df_melt["Valor"] != "No") & 
        (df_melt["Valor"] != 0)
    ]
    
    df_melt["Pregunta"] = cols_rango[0].split("_",1)[1]
    df_melt["Opcion"] = df_melt["Pregunta_Opcion"].apply(lambda x: x.split("_",1)[1])
    
    dfs_long.append(df_melt)

df_long = pd.concat(dfs_long, ignore_index=True)

# ==========================================================
# RESULTADOS
# ==========================================================

print("Tabla opción 1 lista: df_opcion1")
print("Tabla opción 2 lista: df_opcion2")
print("Tabla long lista: df_long")
```

```python
import pandas as pd
from functools import reduce

# 1. Columnas Identificadoras
cols_fijas = [
    'ID', 'START', 'END', 'MUNICIPIO_Seleccione el municipio donde va a aplicar la encuesta', 
    'LOCALIDAD', 'Sexo al nacer', 'Según factura de energía eléctrica estrato', 
    'Rango de edad', 'Edad exacta'
]

# 2. Definición de grupos (Asegúrate de que los nombres de las llaves sean distintos a las columnas originales)
grupos = {
    'Resultado_Actividad': [
        'P621_¿En qué actividad ocupa la mayor parte del tiempo?  ',
        'P622_Desempleado', 'P623_Estudiando', 'P624_Actividades de ocio y/o recreación',
        'P625_Oficios del Hogar', 'P626_Buscando trabajo', 
        'P627_Incapacitado Permanente para Trabajar', 'P628_Otro', 
        'P629_No sabe', 'P6210_No responde', 'P6211_Cual_Normalizada'
    ],
    'Resultado_Delito': [
        'P1171_De las siguientes modalidades delictivas, ¿Cuáles son las tres que más le preocupa?',
        'P1172_Hurto a residencias', 'P1173_Homicidio', 'P1174_Lesiones personales ', 
        'P1175_Hurto de vehículos (o alguna de sus partes)', 'P1176_Venta de drogas', 
        'P1177_Violencia intrafamiliar', 'P1178_Secuestro', 'P1179_Violencia sexual', 
        'P11711_Bloqueos (Obstrucción a vías públicas que afecten el orden público) ', 
        'P11712_Extorsión', 'P11712_Hurto a establecimientos comerciales', 
        'P11713_Vandalismo', 'P11714_Terrorismo', 'P11715_Trata de personas', 
        'P11716_Violencias contra la mujer (física, psicológica o económica)', 
        'P11717_Desaparición forzada', 'P11718_Desplazamiento forzado intraurbano', 
        'P11719_Amenazas', 'P11720_Feminicidio (asesinato de mujeres)', 'P11721_Abigeato'
    ],
    'Resultado_Fuente': [
        'P1281_Con relación a la seguridad de Bogotá ¿Cuáles diría usted que son las tres principales fuentes de información que le permite formar una opinión sobre los temas de seguridad en el municipio?  ',
        'P1282_Experiencias de familiares', 'P1283_Comentarios de la gente (voz a voz).', 
        'P1284_Redes sociales (Twitter, Facebook, grupos de Whatsapp).', 
        'P1285_Noticieros de Televisión.', 'P1286_Revistas (impresas o digitales).', 
        'P1287_Radio.', 'P1288_Periódicos (impresos o digitales).', 
        'P1289_Páginas oficiales en internet (Alcaldía, Fiscalía, Policía, Medicina Legal).'
    ]
}

dfs_procesados = []

for nombre_col, cols_grupo in grupos.items():
    # 3. Verificamos que las columnas existan en el DF antes de procesar
    cols_presentes = [c for c in cols_grupo if c in df_EPV.columns]
    
    if cols_presentes:
        # Transformamos a formato largo
        temp = pd.melt(df_EPV, id_vars=cols_fijas, value_vars=cols_presentes, value_name=nombre_col)
        
        # Eliminamos nulos para que solo queden las opciones marcadas
        temp = temp.dropna(subset=[nombre_col])
        
        # Seleccionamos columnas finales para este grupo
        dfs_procesados.append(temp[cols_fijas + [nombre_col]])

# 4. Unión de todos los dataframes (Merge Outer)
# Esto garantiza que si el ID 1 tiene 2 actividades y 1 delito, salgan 2 filas con ese delito repetido
df_EPV_final = reduce(lambda left, right: pd.merge(left, right, on=cols_fijas, how='outer'), dfs_procesados)

print("¡Proceso completado!")
print(f"Dimensiones finales: {df_EPV_final.shape}")
print(df_EPV_final.head())
```
